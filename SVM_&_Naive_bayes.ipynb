Theory questions

1. What is a Support Vector Machine (SVM)?

Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. It works by finding the optimal hyperplane that separates data points of different classes with the maximum margin. The data points closest to the hyperplane are called support vectors. SVM aims to maximize the distance between the hyperplane and support vectors. It is effective in high-dimensional spaces and works well for both linear and non-linear data using kernel functions.

2. Difference between Hard Margin and Soft Margin SVM

Hard Margin SVM is used when the data is perfectly linearly separable. It does not allow any misclassification and tries to find a strict separating hyperplane. Soft Margin SVM is used when the data is not perfectly separable. It allows some misclassification using slack variables. Soft margin introduces a penalty parameter (C) to control errors. Hard margin may overfit, while soft margin provides better generalization.

3. Mathematical intuition behind SVM

SVM tries to find a hyperplane defined by the equation w·x + b = 0. The goal is to maximize the margin, which is equal to 2/||w||. This is done by minimizing ||w||² subject to classification constraints. It is formulated as an optimization problem. Using Lagrange multipliers, the problem is converted into a dual form. The solution depends only on support vectors.

4. Role of Lagrange Multipliers in SVM

Lagrange multipliers are used to solve the constrained optimization problem in SVM. They help convert the primal problem into its dual form. In the dual form, the solution depends only on inner products of data points. Only non-zero multipliers correspond to support vectors. This makes computation efficient. They also enable the use of the kernel trick.

5. What are Support Vectors in SVM?

Support vectors are the data points that lie closest to the separating hyperplane. They directly influence the position and orientation of the hyperplane. If support vectors are removed, the hyperplane changes. They have non-zero Lagrange multipliers. These points define the margin boundaries. Only support vectors are used in the final decision function.

6. What is Support Vector Classifier (SVC)?

Support Vector Classifier (SVC) is the classification version of SVM. It finds the optimal hyperplane that separates classes with maximum margin. It can handle both linear and non-linear data using kernel functions. SVC is widely used for binary classification problems. It can also be extended to multi-class classification. It performs well in high-dimensional datasets.

7. What is Support Vector Regressor (SVR)?

Support Vector Regressor (SVR) is the regression version of SVM. It tries to fit a function within a certain error margin called epsilon (ε). Instead of minimizing classification error, it minimizes prediction error. It allows some deviations within the ε-tube. SVR focuses on points outside the margin. It is effective for non-linear regression using kernels.

8. What is the Kernel Trick in SVM?

The Kernel Trick is a method used to transform non-linearly separable data into a higher-dimensional space. Instead of computing coordinates directly, it computes inner products using kernel functions. Common kernels include Linear, Polynomial, and RBF. It avoids explicit computation of high-dimensional mapping. This makes SVM efficient. It helps solve complex classification problems.

9. Compare Linear, Polynomial, and RBF Kernel

The Linear Kernel is used when data is linearly separable and is computationally simple. The Polynomial Kernel can model curved decision boundaries and includes a degree parameter. The RBF (Radial Basis Function) Kernel maps data into infinite-dimensional space. RBF is powerful for complex patterns. Linear is faster, while RBF is more flexible. Polynomial lies between them in complexity.

10. Effect of C parameter in SVM

The C parameter controls the trade-off between margin size and classification error. A small C allows more misclassification and gives a wider margin. A large C tries to classify all training points correctly. High C may lead to overfitting. Low C may underfit. Proper tuning of C improves model performance.

11. Role of Gamma in RBF Kernel

Gamma defines how far the influence of a single training example reaches. A high gamma value creates a narrow decision boundary. It may lead to overfitting. A low gamma creates a smoother boundary. It considers points far away in decision making. Proper tuning of gamma is important for performance.

12. What is Naïve Bayes and why is it called “Naïve”?

Naïve Bayes is a probabilistic classification algorithm based on Bayes’ Theorem. It assumes that features are independent given the class label. It calculates posterior probability for each class. The class with highest probability is selected. It is called “Naïve” because of the strong independence assumption. Despite this assumption, it performs well in many tasks.

13. What is Bayes’ Theorem?

Bayes’ Theorem describes the probability of an event based on prior knowledge. It is given by:
P(A|B) = [P(B|A) P(A)] / P(B).
Here, P(A|B) is posterior probability. P(A) is prior probability. P(B|A) is likelihood. It is widely used in classification problems.

14. Difference between Gaussian, Multinomial, and Bernoulli Naïve Bayes

Gaussian Naïve Bayes is used for continuous data assuming normal distribution. Multinomial Naïve Bayes is used for discrete count data like text word frequency. Bernoulli Naïve Bayes is used for binary features (0 or 1). Gaussian uses mean and variance. Multinomial works well for text classification. Bernoulli considers presence or absence of features.

15. When to use Gaussian Naïve Bayes?

Gaussian Naïve Bayes is used when features are continuous. It assumes data follows normal distribution. It is suitable for datasets like medical measurements. It works well when sample size is small. It is simple and fast to implement.

16. Key assumptions of Naïve Bayes

The main assumption is conditional independence of features. It assumes features contribute independently to the class label. It assumes equal importance of features. For Gaussian NB, it assumes normal distribution. These assumptions simplify computation. Though unrealistic, it works effectively in practice.

17. Advantages and Disadvantages of Naïve Bayes

Advantages include simplicity, fast computation, and good performance on large datasets. It works well for text classification. It requires less training data. Disadvantages include strong independence assumption. It may perform poorly when features are correlated. Probability estimates may not be accurate.

18. Why is Naïve Bayes good for text classification?

Text data has high dimensionality. Naïve Bayes handles high-dimensional sparse data well. It works efficiently with word frequencies. It is fast and scalable. It performs well even with small training data. Hence, it is widely used in spam detection and sentiment analysis.

19. Compare SVM and Naïve Bayes

SVM is a margin-based classifier, while Naïve Bayes is probabilistic. SVM works well for complex boundaries. Naïve Bayes is faster and simpler. SVM requires parameter tuning. Naïve Bayes assumes feature independence. SVM generally gives higher accuracy but is computationally expensive.

20. How does Laplace Smoothing help in Naïve Bayes?

Laplace smoothing adds 1 to each feature count to avoid zero probability. Zero probability can make the entire posterior zero. It ensures no class is completely eliminated. It improves model robustness. It is especially useful in text classification. It handles unseen words effectively.

```python
#24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Load dataset (2 features for visualization)
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

# Train model
model = SVC(kernel='poly', degree=3)
model.fit(X, y)

# Create mesh grid
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM with Polynomial Kernel")
plt.show()
```

```python
#26. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20 Newsgroups dataset.
# Import libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Creating a small offline text dataset
texts = [
    "I love programming in python",
    "Python is great for machine learning",
    "Artificial intelligence and machine learning are future technologies",
    "I enjoy watching movies",
    "Movies and music are entertainment",
    "Music is very relaxing",
    "Deep learning is part of artificial intelligence",
    "I like action movies"
]

# Labels: 0 = Technology, 1 = Entertainment
labels = [0, 0, 0, 1, 1, 1, 0, 1]

# Step 2: Convert text into numerical features
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Step 3: Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.3, random_state=42
)

# Step 4: Train Multinomial Naïve Bayes model
model = MultinomialNB()
model.fit(X_train, y_train)

# Step 5: Predict
y_pred = model.predict(X_test)

# Step 6: Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Load dataset
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

C_values = [0.1, 1, 100]

for C in C_values:
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure()
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.title(f"SVM with C = {C}")
    plt.show()
```

```python
#43. Write a Python program to train an SVM Classifier and visualize the Confusion Matrix using seaborn.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
#45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42)

model = GaussianNB()
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]

print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
#46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve

data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42)

model = SVC(kernel='linear', probability=True)
model.fit(X_train, y_train)

y_prob = model.predict_proba(X_test)[:, 1]

precision, recall, _ = precision_recall_curve(y_test, y_prob)

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
